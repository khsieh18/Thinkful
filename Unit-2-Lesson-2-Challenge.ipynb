```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('avocado.csv')
print(df)
```

```python
df_conventional = df[df.type == 'conventional']
df_organic = df[df.type == 'organic']
```

```python
plt.hist([df_conventional['AveragePrice'],df_organic['AveragePrice']], bins=10, rwidth=0.8, color=['green','orange'],
        label=['conventional','organic'])
plt.legend()
plt.xlabel('Average Price')
```

```python
# From the histogram above, we can tell the followings:
# the average conventional avocado price is around 1.1.
# the average organic avocado price is between 1.5 and 1.7.
# the price of conventional avocado is rarely less thant 6.5, and higher than 1.75.
# the price of organic avocado is rarely less than 1.0, and higher than 2.5.
```

```python
plt.scatter(df['AveragePrice'], df['Total Volume'])
plt.xlabel('Average Price')
```

```python
# From the scatter plot above, we can tell the followings:
# the avocados are sold more when the average price is between 1 and 1.5.
```

```python
plt.plot(df['AveragePrice'], df['Total Volume'])
plt.xlabel('Average Price')
```

```python
# From the line plot above, we can tell the followings:
# the avocados are sold more when the average price is between 1 and 1.5.
```

```python
df_region = df.groupby('region').aggregate(np.mean)
df_region
```

```python
plt.boxplot(df_region['AveragePrice'])
```

```python
# from the boxplot above, we can tell the followings:
# the median price of avocado is around 1.4.
# 75% of avocado is sold for less than 1.5.
# 25% of avocado is sold for less than 1.3.
# no avocado is sold for more than 1.9.
# no avocado is sold for less than 0.3.
# there are no outliers in the dataset.
```

```python
df_year = df.groupby('year').aggregate(np.mean)
df_year
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot([2015,2016,2017,2018], df_year['AveragePrice'])
plt.xlabel('Year')
plt.ylabel('Average Price')

plt.subplot(1,2,2)
plt.plot([2015,2016,2017,2018], df_year['Total Volume'])
plt.xlabel('Year')
plt.ylabel('Total Volume')
```

```python
# From the line plot above, we can tell the followings:
# when the average price increases, the total sales volume stays pretty much the same.
# when the average price decreases, the total sales volume increases.
```
