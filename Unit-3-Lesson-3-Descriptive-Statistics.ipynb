```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rand1 = np.random.binomial(50,0.5,100)
rand2 = np.random.exponential(50,100)
rand3 = np.random.gamma(2,1,100)
rand4 = np.random.normal(150,10,100)
rand5 = np.random.poisson(5,100)
rand6 = np.random.uniform(0,1,100)
```

```python
plt.figure(figsize=(10,10))

plt.subplot(2,3,1)
plt.hist(rand1)
mean=np.mean(rand1)
std=np.std(rand1)
plt.axvline(x=mean, color='black')
plt.axvline(x=mean+std, color='red')
plt.axvline(x=mean-std, color='red')
plt.title('binomial distribution')

plt.subplot(2,3,2)
plt.hist(rand2)
mean=np.mean(rand2)
std=np.std(rand2)
plt.axvline(x=mean, color='black')
plt.axvline(x=mean+std, color='red')
plt.axvline(x=mean-std, color='red')
plt.title('exponential distribution')

plt.subplot(2,3,3)
plt.hist(rand3)
mean=np.mean(rand3)
std=np.std(rand3)
plt.axvline(x=mean, color='black')
plt.axvline(x=mean+std, color='red')
plt.axvline(x=mean-std, color='red')
plt.title('gamma distribution')

plt.subplot(2,3,4)
plt.hist(rand4)
mean=np.mean(rand4)
std=np.std(rand4)
plt.axvline(x=mean, color='black')
plt.axvline(x=mean+std, color='red')
plt.axvline(x=mean-std, color='red')
plt.title('normal distribution')

plt.subplot(2,3,5)
plt.hist(rand5)
mean=np.mean(rand5)
std=np.std(rand5)
plt.axvline(x=mean, color='black')
plt.axvline(x=mean+std, color='red')
plt.axvline(x=mean-std, color='red')
plt.title('poisson distribution')

plt.subplot(2,3,6)
plt.hist(rand6)
mean=np.mean(rand6)
std=np.std(rand6)
plt.axvline(x=mean, color='black')
plt.axvline(x=mean+std, color='red')
plt.axvline(x=mean-std, color='red')
plt.title('uniform distribution')
```
